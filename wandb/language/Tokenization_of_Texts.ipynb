```python
# Group 1
from nltk.tokenize import RegexpTokenizer

def Reference():
  file1 = open('references1.txt', 'r')
  Lines = file1.readlines()
  tokenizer = RegexpTokenizer(r'\w+')
  count = 0
  tokenized_sents = [tokenizer.tokenize(line) for line in Lines]
  for line in tokenized_sents:
    count += 1
    reference = [line]
    print("reference{} = {}".format(count,reference))
def Candidate():
  file1 = open('MAL1.txt', 'r')
  Lines = file1.readlines()
  tokenizer = RegexpTokenizer(r'\w+')
  count = 0
  tokenized_sents = [tokenizer.tokenize(line) for line in Lines]
  for line in tokenized_sents:
    count += 1
    candidates = line
    print("Candidate{} = {}".format(count,candidates))
def Candidate2():
  file1 = open('Seq2Seq1.txt', 'r')
  Lines = file1.readlines()
  tokenizer = RegexpTokenizer(r'\w+')
  count = 0
  tokenized_sents = [tokenizer.tokenize(line) for line in Lines]
  for line in tokenized_sents:
    count += 1
    candidates2 = line
    print("Candidate2{} = {}".format(count,candidates2))
print('--------------------------------------------------------------------------------------------------------------------------------')
Reference()
print('--------------------------------------------------------------------------------------------------------------------------------')
Candidate()
print('--------------------------------------------------------------------------------------------------------------------------------')
Candidate2()
print('--------------------------------------------------------------------------------------------------------------------------------')
```
